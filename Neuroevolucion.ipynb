```python
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input

# =========================
# DATASET
# =========================
X, y = make_classification(n_samples=400, n_features=20, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)

# =========================
# REPRESENTACIÓN
# =========================
def create_individual():
    num_layers = random.randint(1, 3)
    return [random.randint(10, 100) for _ in range(num_layers)]

# =========================
# MODELO
# =========================
def build_model(chromosome):
    model = Sequential()

    # Capa de entrada (forma correcta)
    model.add(Input(shape=(20,)))

    # Primera capa oculta
    model.add(Dense(chromosome[0], activation='relu'))

    # Capas ocultas adicionales
    for n in chromosome[1:]:
        model.add(Dense(n, activation='relu'))

    # Capa de salida
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model

# =========================
# FITNESS
# =========================
def fitness(individual):
    model = build_model(individual)
    model.fit(X_train, y_train, epochs=3, batch_size=16, verbose=0)
    y_pred = (model.predict(X_val, verbose=0) > 0.5).astype(int)
    return accuracy_score(y_val, y_pred)

# =========================
# SELECCIÓN
# =========================
def selection(population, scores):
    i, j = random.sample(range(len(population)), 2)
    return population[i] if scores[i] > scores[j] else population[j]

# =========================
# CRUZAMIENTO
# =========================
def crossover(p1, p2):
    min_len = min(len(p1), len(p2))
    if min_len < 2:
        return p1.copy()
    cut = random.randint(1, min_len - 1)
    return p1[:cut] + p2[cut:]

# =========================
# MUTACIÓN
# =========================
def mutation(ind):
    if random.random() < 0.3:
        idx = random.randint(0, len(ind)-1)
        ind[idx] = random.randint(10, 100)
    return ind

# =========================
# VISUALIZACIÓN POBLACIÓN
# =========================
def print_population(pop, scores, gen):
    print(f"\n===== GENERACIÓN {gen} =====")
    print("ID | Cromosoma        | Fitness")
    print("--------------------------------")
    for i, (ind, sc) in enumerate(zip(pop, scores)):
        print(f"{i:<2} | {str(ind):<15} | {sc:.4f}")

# =========================
# ALGORITMO GENÉTICO
# =========================
pop_size = 5
generations = 8

population = [create_individual() for _ in range(pop_size)]

best_history = []
avg_history = []

plt.ion()

for gen in range(1, generations+1):

    # Evaluación
    scores = [fitness(ind) for ind in population]

    print_population(population, scores, gen)

    best_history.append(max(scores))
    avg_history.append(np.mean(scores))

    # Nueva población
    new_pop = []

    print("\n--- REPRODUCCIÓN ---")

    while len(new_pop) < pop_size:
        p1 = selection(population, scores)
        p2 = selection(population, scores)

        print(f"\nPadre 1: {p1}")
        print(f"Padre 2: {p2}")

        # Cruzamiento
        child = crossover(p1, p2)
        print(f"Hijo (crossover): {child}")

        # Mutación
        child = mutation(child)
        print(f"Hijo (mutado): {child}")

        new_pop.append(child)

    population = new_pop

    # =========================
    # GRÁFICA EN TIEMPO REAL
    # =========================
    plt.clf()
    plt.plot(best_history, label="Mejor fitness")
    plt.plot(avg_history, label="Promedio fitness")
    plt.xlabel("Generación")
    plt.ylabel("Fitness")
    plt.title("Evolución del Algoritmo Genético")
    plt.legend()
    plt.pause(0.5)

plt.ioff()
plt.show()
```

```python
from tensorflow.keras import backend as K

def fitness(individual, verbose_metrics=False):
    K.clear_session()

    model = build_model(individual)

    history = model.fit(
        X_train, y_train,
        epochs=5,
        batch_size=16,
        validation_data=(X_val, y_val),
        verbose=0
    )

    # Predicción final
    y_pred = (model.predict(X_val, verbose=0) > 0.5).astype(int)
    acc = accuracy_score(y_val, y_pred)

    # 🔍 Mostrar métricas si se desea
    if verbose_metrics:
        print("\nEntrenamiento del individuo:", individual)
        print("Epoch | Train Acc | Val Acc | Train Loss | Val Loss")

        for i in range(len(history.history['accuracy'])):
            print(f"{i+1:5} | "
                  f"{history.history['accuracy'][i]:.4f} | "
                  f"{history.history['val_accuracy'][i]:.4f} | "
                  f"{history.history['loss'][i]:.4f} | "
                  f"{history.history['val_loss'][i]:.4f}")

    return acc, history.history

scores = []
histories = []

for ind in population:
    score, hist = fitness(ind)
    scores.append(score)
    histories.append(hist)
```

```python
best_idx = np.argmax(scores)
best_history = histories[best_idx]

plt.figure(figsize=(10,4))

# Accuracy
plt.subplot(1,2,1)
plt.plot(best_history['accuracy'], label='Train Acc')
plt.plot(best_history['val_accuracy'], label='Val Acc')
plt.title(f'Generación {gen} - Accuracy')
plt.legend()

# Loss
plt.subplot(1,2,2)
plt.plot(best_history['loss'], label='Train Loss')
plt.plot(best_history['val_loss'], label='Val Loss')
plt.title(f'Generación {gen} - Loss')
plt.legend()

plt.show()
```
